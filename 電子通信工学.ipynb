```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
%matplotlib inline
```

```python
x = np.linspace(0, 1000, 100000)
```

```python
R = 1
L = 0.001
C = 0.01
```

```python
def expr(R,L,C,w):
    return R/np.sqrt(((w*L)/((1-w**2*L*C))**2+R**2))
```

```python
plt.plot(x,expr(R,L,C,x))
```

```python
def expr(R,L,C,w):
    return R/np.sqrt(R**2 + ((1-w**2*L*C)/(w*C))**2)
```

```python
x = np.linspace(0, 2000, 100000)
plt.plot(x,expr(R,L,C,x))
```
