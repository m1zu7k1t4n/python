```python
%matplotlib inline
import math
import numpy as np
import matplotlib.pyplot as plt
from pylab import rcParams
from ipywidgets import interact
import ipywidgets as widgets
```

```python
def waveplot(wavelist):
    temp = wavelist[0] + wavelist[1]
    for wave in wavelist[2:]:
        temp += wave
    plt.plot(time,temp)
```

```python
rcParams['figure.figsize'] = 10,10
time = np.linspace(0,10,1000)
omega_0 = 1
plt.figure()
plt.subplots_adjust(wspace=0.5,hspace=0.5)
```

```python
@interact(leng=widgets.IntSlider(min=6,max=100000,step=10,value=100))
def square(leng):
    wavelist = list()
    for i in range(1,leng,2):
        temp = (1/i)*np.sin(i * omega_0 * time)
        wavelist.append(temp)
    waveplot(wavelist)
```
