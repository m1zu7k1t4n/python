```python
def rgb_output():
    r = 0
    g = 0
    b = 0
    while True:
        for i in range(6):
            r = i
            for v in range(6):
                g = v
                for k in range(6):
                    b = k
                    yield (r, g, b)
```

```python
g = rgb_output()
```

```python
results = []
for i in range(220):
    results.append(g.__next__())
```
